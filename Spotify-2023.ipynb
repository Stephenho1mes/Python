```python
#The Average Danceability across all tracks

avg_danceability = spotify_2023['danceability_%'].mean()
print(f"Average danceability: {avg_danceability:.2f}%")
```

```python
top3_artists = spotify_2023[['artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(3)

print(f"Top 3 Artists by Total Streams:\n{top3_artists}\n")
```

```python
highest_danceability_track = spotify_2023[spotify_2023['danceability_%'] == spotify_2023['danceability_%'].max()]['track_name'].values[0]
lowest_danceability_track = spotify_2023[spotify_2023['danceability_%'] == spotify_2023['danceability_%'].min()]['track_name'].values[0]

print(f"Highest Danceability Track: {highest_danceability_track}")
print(f"Lowest Danceability Track: {lowest_danceability_track}")
```

```python
spotify_2023 = spotify_2023.sort_values(by="streams", ascending=True).head(10)


plt.figure(figsize=(10, 8))
plt.barh(spotify_2023['track_name'], spotify_2023['streams'])
plt.xlabel('Number of Streams')
plt.ylabel('Track Name')
plt.title('Number of Streams per Track')
plt.tight_layout()
plt.show()

#How do I tidy the bottom section up?
```

```python
#Pie chart provides an overview on what key the music is played in.  Major being moree popular

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

spotify_2023['mode'].value_counts().plot(kind='pie', ax=axes[0])

spotify_2023['key'].value_counts().plot(kind='pie', ax=axes[1])

axes[0].set_ylabel('')
axes[1].set_ylabel('')
fig.suptitle('Music Note')
plt.tight_layout()
```
